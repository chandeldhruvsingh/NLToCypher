```python
def extract_questions_and_cyphers(input_file_path, output_file_path):
    with open(input_file_path, 'r') as file:
        data = json.load(file)

    extracted_data = [
        {"Question": item["Question"], "Cypher": item["Cypher"]}
        for item in data[:1000]
    ]

    # Saving the extracted data to a new JSON file
    with open(output_file_path, 'w') as output_file:
        json.dump(extracted_data, output_file, indent=4)
```
