```python
import json



```

```python
# Load the JSON data
with open('/Users/dhruvchandel/VSCodeProjects/NLToSPARQL/dataset/qald-9-train-multilingual.json', 'r') as f:
    data = json.load(f)
```

```python
# Iterate through each question
for question in data['questions']:
    # Check if the question is in English
    for q in question['question']:
        if q['language'] == 'en':
            english_question = q['string']
            break

    # Get the SPARQL query
    sparql_query = question['query']['sparql']

    # Get all the results
    results = []
    for answer in question['answers']:
        if 'results' in answer and 'bindings' in answer['results']:
            for result in answer['results']['bindings']:
                if 'uri' in result:
                    results.append(result['uri']['value'])

    # Print the English question, SPARQL query, and results
    print("English Question:", english_question)
    print("SPARQL Query:", sparql_query)
    print("Results:")
    for result in results:
        print(result)
    print()
```
